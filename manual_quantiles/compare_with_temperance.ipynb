```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

from utils import to_GeV_per_cubic_femtometre
```

```python
# Temperance 
x = to_GeV_per_cubic_femtometre(np.linspace(3e13, 5e15, 100))
quantiles = to_GeV_per_cubic_femtometre(pd.read_csv('quantiles/p_of_eps_quantiles.csv'))
quantiles_ns10_Xmarg = to_GeV_per_cubic_femtometre(pd.read_csv('quantiles/p_of_eps_quantiles_ns10_Xmarg.csv'))

# My data
energy_density_grid = np.linspace(1e-10, 5, 1000)
weighted_quantiles = np.loadtxt('weighted_quantiles/weighted_quantiles_p_e.dat')
weighted_quantiles_combined_10 = np.loadtxt('weighted_quantiles/weighted_quantiles_combined_p_e_10.dat')
```

```python
fig, ax = plt.subplots()

ax.set_yscale('log')

ax.fill_between(
    energy_density_grid, 
    weighted_quantiles[0], 
    weighted_quantiles[2], 
    alpha=0.25, 
    color='k', 
    edgecolor=None,
    )

ax.plot(x, quantiles.loc[5,:].values)
ax.plot(x, quantiles.loc[50,:].values)
ax.plot(x, quantiles.loc[95,:].values)

ax.set_ylim(1e-3, 1)
ax.set_xlim(0., 2)
```

```python
fig, ax = plt.subplots()

ax.set_yscale('log')

ax.fill_between(
    energy_density_grid, 
    weighted_quantiles_combined_10[0], 
    weighted_quantiles_combined_10[2], 
    alpha=0.25, 
    color='k', 
    edgecolor=None,
    )

ax.plot(x, quantiles_ns10_Xmarg.loc[5,:].values)
ax.plot(x, quantiles_ns10_Xmarg.loc[50,:].values)
ax.plot(x, quantiles_ns10_Xmarg.loc[95,:].values)

ax.set_ylim(1e-3, 1)
ax.set_xlim(0., 2)
```
