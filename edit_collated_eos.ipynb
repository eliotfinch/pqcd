```python
import numpy as np
import pandas as pd
```

```python
# Load the collated EoSs and filter out the ones with zero weight
collated_eos = pd.read_csv('collated_np_all_post.csv')
nonzero_collated_eos = collated_eos[collated_eos.logweight_total > -np.inf]
```

```python
nonzero_collated_eos
```

```python
# The pre-computed weights of these EoSs
weights = np.exp(nonzero_collated_eos.logweight_total.values)
```

```python
for ns in [7, 10]:

    qcd_weights = np.loadtxt(f'weights/qcd_weights_{ns:02}.dat')
    combined_weights = weights*qcd_weights

    # Add the combined weights to the collated EoSs
    nonzero_collated_eos[f'weight_ns{ns:02}_Xmarg'] = combined_weights

# Save the collated EoSs with the combined weights
nonzero_collated_eos.to_csv('collated_eos.csv', index=False)
```
