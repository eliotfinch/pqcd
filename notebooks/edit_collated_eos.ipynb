```python
import numpy as np
import pandas as pd
```

```python
# Load the collated EOSs
collated_eos = pd.read_csv('../data/eos-draws-default/eos-draws-default.csv')
```

```python
collated_eos
```

```python
# The maximised likelihood weights
# --------------------------------

# Fixed nterm
for nterm in [2, 3, 4, 5, 6, 7, 8, 9, 10]:

        
    # Marginalized over X
    pqcd_weights = np.loadtxt(f'../data/eos-draws-default/pqcd-weights/pqcd_weights_{nterm:02}nsat_Xmarg_mu2.6.dat')

    # Add the weights to the collated EOSs
    collated_eos.insert(len(collated_eos.columns), f'pqcd_weight_{nterm:02}nsat_Xmarg_mu2.6', pqcd_weights)

    # Fixed X
    for X in [0.5, 2]:
        pqcd_weights = np.loadtxt(f'../data/eos-draws-default/pqcd-weights/pqcd_weights_{nterm:02}nsat_X{X}_mu2.6.dat')

        # Add the weights to the collated EOSs
        collated_eos.insert(len(collated_eos.columns), f'pqcd_weight_{nterm:02}nsat_X{X}_mu2.6', pqcd_weights)

# nterm = nTOV

pqcd_weights = np.loadtxt('../data/eos-draws-default/pqcd-weights/pqcd_weights_ntov_Xmarg_mu2.6.dat')
collated_eos.insert(len(collated_eos.columns), 'pqcd_weight_ntov_Xmarg_mu2.6', pqcd_weights)

for X in [0.5, 2]:
    pqcd_weights = np.loadtxt(f'../data/eos-draws-default/pqcd-weights/pqcd_weights_ntov_X{X}_mu2.6.dat')
    collated_eos.insert(len(collated_eos.columns), f'pqcd_weight_ntov_X{X}_mu2.6', pqcd_weights)

# The marginalized likelihood weights
# -----------------------------------

for nterm in [2, 3, 4, 5, 6, 7, 8, 9, 10]:
    pqcd_weights_marg = np.loadtxt(f'../data/eos-draws-default/pqcd-weights/pqcd_weights_{nterm:02}nsat_marg.dat')
    collated_eos.insert(len(collated_eos.columns), f'pqcd_weight_{nterm:02}nsat_marg', pqcd_weights_marg)
        
# nterm = nTOV
pqcd_weights_marg = np.loadtxt('../data/eos-draws-default/pqcd-weights/pqcd_weights_ntov_marg.dat')
collated_eos.insert(len(collated_eos.columns), 'pqcd_weight_ntov_marg', pqcd_weights_marg)
```

```python
collated_eos.to_csv('../data/eos-draws-default/eos-draws-default.csv', index=False)
```
