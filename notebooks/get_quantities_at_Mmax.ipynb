```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from pathlib import Path
```

```python
set_number = 12
eos_dir = Path(f'../data/eos-draws-modified/{set_number:02}/margagn')
```

```python
eos_draws = pd.read_csv(
    f'../data/eos-draws-modified-{set_number:02}.csv', 
    index_col='eos'
)
```

```python
eos_draws
```

```python
rhoc_at_Mmax = []
for eos, entry in eos_draws.iterrows():
    Mmax = entry.Mmax
    macro_draw = pd.read_csv(
        eos_dir / f'DRAWmod1000-{int(eos/1000):06}/macro-eos-draw-{eos:06}.csv'
    )
    Mmax_index = np.argmin((macro_draw.M - Mmax)**2)
    rhoc_at_Mmax.append(macro_draw.central_baryon_density[Mmax_index])
```

```python
eos_draws.insert(1, 'rhoc(M@Mmax)', rhoc_at_Mmax)
```

```python
eos_draws
```

```python
eos_draws.to_csv(f'../data/eos-draws-modified-{set_number:02}.csv')
```
